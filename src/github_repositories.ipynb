```python
# Ambient config
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.style.use('classic')
plt.rcParams['figure.figsize'] = (20,8)

# Loading file
df = pd.read_csv(r'./github_repositories.csv', on_bad_lines = 'skip')

# Data example
df.sample(3)
```

```python
# Repositorios com mais 'Stars'
df.loc[0:10, ['Name','Stars', 'Forks', 'Language', 'Description', 'Growth Pattern']].sort_values(by = 'Stars', ascending = False)
```

```python
# Repositorios com mais 'Forks'
df.loc[0:10, ['Name','Stars', 'Forks', 'Language', 'Description']].sort_values(by = 'Forks', ascending = False)
```

```python
xticks = np.arange(0, 2000, 50)

df.groupby('Language').size().sort_values(ascending = True).tail(20).plot(kind = 'barh', title = 'Linguagens mais utilizadas', xlabel = 'Quantidade de Repositorios',xticks=xticks).margins(x=0.1)
```
